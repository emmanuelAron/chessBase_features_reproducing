```python
#Parsing:
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, "html.parser")
type(soup)
print(soup.prettify())  # This formats the HTML in a readable way
```

```python
# Search by tag
soup.find_all("title") # Find the all <title> tags on the page
```

```python
# <table class="table"> : all my datas are inside this html <table>
result = soup.find(class_='table')
result
```

```python
result
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Use 'result' directly, assuming it contains the HTML of the table
soup = BeautifulSoup(str(result), 'html.parser')  # Convert result to string if necessary

# Lists to store extracted data
eco_codes = []
openings = []
moves_list = []

# Find all components (table rows, excluding the header)
components = soup.find_all('tr')[1:]  # Skip the header

for component in components:
    cells = component.find_all('td')
    if len(cells) == 3:
        # Extract ECO code
        eco_code = cells[0].text.strip()
        eco_codes.append(eco_code)

        # Extract opening name
        opening_name = cells[1].text.strip()
        openings.append(opening_name)

        # Extract moves
        moves = cells[2].text.strip()
        moves_list.append(moves)

# Create DataFrame
df = pd.DataFrame({
    'ECO_Code': eco_codes,
    'Opening': openings,
    'Moves': moves_list
})

df


```

```python
#pd.set_option('display.max_rows', None)
df
```

```python
import re
# Function to clean up move numbering
def remove_numbering(moves):
    # Remove any move numbers (e.g., "1.", "2.", etc.)
    clean_moves = re.sub(r'\d+\.', '', moves)
    # Remove any extra whitespace left by removing numbers
    clean_moves = ' '.join(clean_moves.split())
    return clean_moves

# Apply the cleaning function to the 'Moves' column
df['Moves'] = df['Moves'].apply(remove_numbering)

# Display the cleaned DataFrame
display(df)
```

```python
import pandas as pd
import re
from IPython.display import display

# Function to add numbering to each move pair
def add_move_numbering(moves):
    # Split the moves into individual half-moves
    moves_list = moves.split()
    formatted_moves = []
    move_number = 1
    
    # Iterate through moves in pairs
    for i in range(0, len(moves_list), 2):
        if i + 1 < len(moves_list):
            # Format as "1. move1 move2" if there are two moves in the pair
            formatted_moves.append(f"{move_number}. {moves_list[i]} {moves_list[i + 1]}")
        else:
            # Format as "1. move1" if there's only one move left
            formatted_moves.append(f"{move_number}. {moves_list[i]}")
        move_number += 1

    # Join the formatted moves back into a single string
    return ' '.join(formatted_moves)

# Apply the numbering function to the 'Moves' column
df['Moves'] = df['Moves'].apply(add_move_numbering)

# Display the DataFrame with numbered moves
display(df)
```

```python
# Conversion to csv:
df.to_csv("ECO_code_mapping.csv")
```
