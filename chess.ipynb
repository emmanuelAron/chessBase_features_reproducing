```python
import chess
import chess.svg
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image

# Create the board
board = chess.Board()

# Add moves
moves = ["d4", "d5", "c4", "e6"]
for move in moves:
    board.push_san(move)

# Convert the position to an image
svg_image = chess.svg.board(board=board)
image = Image.open(BytesIO(svg_image.encode('utf-8')))

# Display the image with Matplotlib
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
import chess
import chess.svg
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image
import cairosvg

# Create the board
board = chess.Board()

# Add moves
moves = ["d4", "d5", "c4", "e6"]
for move in moves:
    board.push_san(move)

# Convert the SVG to a PNG using cairosvg
svg_image = chess.svg.board(board=board)
png_image = BytesIO()
cairosvg.svg2png(bytestring=svg_image.encode('utf-8'), write_to=png_image)

# Load the PNG image with PIL
png_image.seek(0)
image = Image.open(png_image)

# Display the image with Matplotlib
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis('off')
plt.show()
```
